# Simple OpenCV method

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib
```

```python
image = cv2.imread('/home/orelit/Projects -Sathsara/Planigo/data/test/7290015350143/7290015350143 (2).jpeg', 
                   flags=cv2.IMREAD_UNCHANGED)
RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
type(RGB_image)
```

```python
plt.imshow(RGB_image)
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
RGB_image_sharp = cv2.cvtColor(image_sharp, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(121), plt.imshow(RGB_image)
plt.subplot(122), plt.imshow(RGB_image_sharp)
  
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16,10)
```
